```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')

tips.head(10)
```

```python
# Types of plots

# Scatter Plot - will be displayed as various points.
# To display the relationship between two continuous variables.
# It's best used for showing correlation or trends.

# sns.scatterplot(x, y, data=None, hue=None, style=None, size=None, palette=None)
# Key Parameters:

# x, y: These are the variables you want to plot on the x-axis and y-axis of the scatter plot. (column names)
# data: The data you are working on
# hue: Categorical column which help us catagorize data into various groups and color them for better visualization. Different colors will be assigned to various categories based on our pallet.
# style: Categorical variable to differentiate markers' style (e.g., dots vs. crosses). Is used to scatter various points via various marker shapes.
# size: Categorical or continuous variable to scale the size of the points. When the size becomes big the dot in our graph will also appear big.
# palette: Color palette or predefined category-color map.
# markers : plot will display actual markers (dots, shapes) for the data points.
# edgecolor: Changes the color of the edges of the markers (data points).

sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day", size="size", palette="Set1", style = "sex", markers=True, edgecolor=None, legend = "brief")
plt.show()
```

```python
# Types of plots

# Line Ploat - To visualize the trend or change of a variable over time or ordered categories.
# Best for continuous variables showing trends.

# sns.lineplot(x, y, data=None, hue=None, style=None, markers=None, dashes=None)

# Key Parameters:
# dashes: Specify which lines should be dashed.

sns.lineplot(x="time", y="tip", data=tips, hue="day", markers=True, dashes = True)
plt.show()
```

```python
# Types of plots

# Bar plot - To compare the means of a categorical variable or group-wise numerical values.
# Best for when you want to summarize the central tendency of a numerical feature per group.

# sns.barplot(x=None, y=None, data=None, hue=None, estimator=None, ci=None)

# Key Parameters:

# estimator: The statistical function to apply on y to summarize (e.g., np.mean, np.sum).
# ci: Confidence intervals for the plot; None for no intervals.


sns.barplot(x="sex", y="tip", hue="day" , data=tips, estimator=np.sum)
plt.show()
```

```python
# Types of plots

# Box plot - To display the distribution of a continuous variable, highlighting the median, quartiles, and potential outliers.
# Ideal for comparing distributions of a variable across several categories.

# sns.boxplot(x=None, y=None, data=None, hue=None, whis=None, fliersize=None)

# Key Parameters:

# whis: Controls the range for outliers (default is 1.5).
# fliersize: Size of points representing outliers.

sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", whis=2,fliersize=1)
plt.show()
```

```python
# Types of plots

# Violin Plot - Similar to boxplots but with a more detailed visualization of the distribution, displaying the density of the data.
# Ideal for comparing distributions across categories, with an added benefit of showing multimodal distributions.

# sns.violinplot(x=None, y=None, data=None, hue=None, split=False, inner="stick")

# Key Parameters:

# split: If True, separate violins for the hue levels; otherwise, a single violin is plotted.
# inner: Visual elements to show inside the violin (e.g., "box", "stick").

sns.violinplot(x="sex", y="total_bill", data=tips, hue="sex", split=True, inner="stick")
plt.show()
```

```python
# Types of plots

# Histograms - For showing the frequency distribution of a single continuous variable.
# It helps to identify the underlying distribution of a data set (e.g., normal, bimodal).

# sns.histplot(data=None, bins=None, kde=False, hue=None, stat="count", discrete=False)
# Key Parameters:

# bins: Number of bins or custom bin edges.
# kde: If True, overlays a Kernel Density Estimate (smoothed curve) on the histogram.
# stat: Option for scaling the histogram (e.g., "count", "density", etc.).

sns.histplot(data=tips, x="total_bill", bins=10, kde=True, hue="sex", stat="probability")
plt.show()
```

```python
# Types of plots
# Heat Map - For displaying matrix data where values in a 2D grid are represented with color.
# Used to visualize correlation matrices or to compare values across two dimensions.

# sns.heatmap(data, annot=False, cmap="coolwarm", cbar=True, fmt=".2f")

# Key Parameters:
# annot: Whether to display the values in each cell.
# cmap: The color map (e.g., "coolwarm", "viridis").
# cbar: Whether to display the color bar.
# fmt: String for formatting numbers (e.g., .2f for two decimal places).

correlation_matrix = tips.select_dtypes(include=["number"]).corr()

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", cbar=True, fmt=".2f")
plt.show()
```

```python
# Types of plots
# Pair plots - Used to visualize pairwise relationships between several continuous variables in your dataset.
# Useful for exploring multivariate datasets and spotting correlations.

# sns.pairplot(data, hue=None, vars=None, kind="scatter")

# Key Parameters:
# vars: Subset of variables to plot pairwise plots for; if None, all variables are used.
# kind: Type of plot ("scatter", "reg", "kde").

sns.pairplot(tips, hue="sex")
plt.show()

```
