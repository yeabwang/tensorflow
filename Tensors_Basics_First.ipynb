Population Mean - Its a mean that can be taken from the entire population data.

Estimated Mean - Since we mostly don't have the full data set we calculate the mean based on the already available data which gives us the estimated mean.

To calculate this we will sum up all the data and divide it with the number of data we have.

The mean tells us the height of the peak. Now to calculate the width of the peak:

Population variance: This tells how the data is scattred around the dataset, the less the variance means the lower the difference between the values in the dataset.

To calculate this we deduct the mean value from each data points, square the out put and find the total sum across the entire dataset. Ones we have that we will divide it with the number of data points we have.

Estimated Variance: This tells us how the data is scattered from our sample dataset.

To calculate this we deduct the each data points from the estimated mean, square it and find the total sum. After we did that we deduct one from the total number of data points we have and divide the total sum with this.

Since variance is a squared value we can't really apply it to the actual data so we take the square root of the vairance and that will give us the standard diviation. How the data is spread out across the dataset.

If we take the population variance it will become a calculated(population standard diviation). If we take the esitmated variance it will become esimated standard deviation.

UNIFORM DISTRIBUTION
When we are talking about uniform distribution, we are saying the values in between of our min and max set values have an equal chance of getting selected, unlike the normal distribution.

```
tf_uniform = tf.random.uniform(
    [3,2],
    minval = 0,
    maxval = 100, #if not set the max value is set to 1
    dtype =tf.dtypes.int32,
    seed = None, #Seed is the actual origin of where the measurement starts, by using similar seed we can create a replicatable results across multiple function calls.
    name = None
)

print(tf_uniform)
```

```
tf.random.set_seed(5)

print(tf.random.uniform(shape = [3,], maxval=3, dtype=tf.int32,seed=10))
print(tf.random.uniform(shape = [3,], maxval=3, dtype=tf.int32,seed=10))
print(tf.random.uniform(shape = [3,], maxval=3, dtype=tf.int32,seed=10))
print(tf.random.uniform(shape = [3,], maxval=3, dtype=tf.int32,seed=10))
```
